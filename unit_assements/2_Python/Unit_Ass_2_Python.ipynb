```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# files needed
file1 = 'resources/avg_rain_state.csv'
file2 = 'resources/donors2008.csv'
```

```python
avg_rain_state_df = pd.read_csv(file1)
donors_df = pd.read_csv(file2)

avg_rain_state_df
# donors_df.head()
```

```python
# Q: bar graph question for avg_rain_state.csv, fill in blanks
# A: 

tick_locations = [value for value in avg_rain_state_df['State']]
plt.figure(figsize=(20,3))
plt.bar(avg_rain_state_df['State'], avg_rain_state_df['Inches'] )
plt.xticks(tick_locations, avg_rain_state_df['State'], rotation = 'vertical')
plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain (inches)")
plt.show()
```

```python
# Q: groupby question, How many states have averages equal to or greater than $750.00?
# A:

donors_avg = donors_df.groupby(['State']).mean()
donors_avg_greaterequal_750 = donors_avg[donors_avg['Amount'] >= 750]
donors_avg_greaterequal_750.count()
```
